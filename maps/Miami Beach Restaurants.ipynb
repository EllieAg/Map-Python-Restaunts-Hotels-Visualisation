```python
import folium
import pandas as pd
```

```python
rest = pd. read_csv ('google_maps.csv')
rest.head(58)
```

```python
mia_map = folium. Map(
    location=[25.761681, -80.191788],
    zoom_start=12
)
for _, r in rest.iterrows():
    folium. Marker(
        location=[r['latitude'], r['longitude']],
        popup=r['Review_Points'],
        tooltip=r['Title'],
        icon=folium.Icon(color="orange", icon="cutlery") 
    ).add_to (mia_map)              
mia_map
```
