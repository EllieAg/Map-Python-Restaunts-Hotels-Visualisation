```python
import folium
import pandas as pd
```

```python
htl = pd.read_csv ('Hotel_Motel_Inn.csv')
htl.head(558)
```

```python
airbnb = pd.read_csv ('Hotel_Motel_Inn.csv')
airbnb.head(226031)
```

```python
mia_map = folium. Map(
    location=[25.761681, -80.191788],
    zoom_start=12
)
for _, h in htl.iterrows():
    folium. Marker(
        location=[h['LAT'], h['LON']],
        popup=h['CAPACITY'],
        tooltip=h['NAME'],
        icon=folium.Icon(color="orange", icon="home") 
    ).add_to (mia_map)
for _, a in airbnb.iterrows():
    folium. Marker(
        location=[a['LAT'], a['LON']],
        popup=a['NAME'],
        tooltip=a['name'],
        icon=folium.Icon(color="blue", icon="home") 
    ).add_to (mia_map)
mia_map
```
